# Part 1: Machine Learning
## Data Preparation:

```python
file_name = "../../ml_datasets/clothing_store.csv"
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv(file_name)
```

```python
data
```

There is a lot of numeric data.  My options are: Logistic Regression, Support Vector Machines, Decision Trees, k-Nearest Neighbors, and Naive Bayes.

I choose: logistic regression and SVM

1. Remove id 
2. Change valphon into 1's and 0's
3. Transform zipcode into a categorical value from 0-9 to represent the 10 regions of the US
    (use one hot encoding)


```python
data = data.drop(['CUSTOMER_ID'], axis=1)
data['VALPHON'] = data['VALPHON'].eq('Y').mul(1)
```

```python
data.loc[data['ZIP_CODE'] <= 9999, 'ZIP_CODE'] = 0
data['ZIP_CODE'] = (data['ZIP_CODE'] / 10000).astype(int)
```

```python
#from sklearn.preprocessing import OneHotEncoder

#enc = OneHotEncoder(handle_unknown='ignore')
#enc.fit(data['ZIP_CODE'])
#data['ZIP_CODE'] = enc.transform(data['ZIP_CODE'])
# df = pd.concat([df,pd.get_dummies(df['mycol'], prefix='mycol',dummy_na=True)],axis=1).drop(['mycol'],axis=1)
data = pd.concat([data, pd.get_dummies(data["ZIP_CODE"],prefix='ZIP_CODE',dummy_na=True)],axis=1).drop(['ZIP_CODE'],axis=1).drop(['ZIP_CODE_nan'],axis=1)
data
```

## Comparing classifers:

Using 30-fold cross validation

```python
data_rows = data.drop(['RESP'], axis=1).to_numpy().tolist()
resp = data['RESP'].to_numpy().tolist()
```

### Logistic Regression

```python
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
import warnings; warnings.simplefilter('ignore')

algorithms = ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'] # default='lbfgs'
logreg = linear_model.LogisticRegression(solver=algorithms[4], random_state = 42)

logreg_cv_scores = cross_val_score(logreg, data_rows, resp, cv=30)
logreg_acc = np.mean(logreg_cv_scores)
logreg_cv_scores
```

### SVM

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf', C=1E3)
svm_cv_scores = cross_val_score(svm, data_rows, resp, cv=30)
svm_acc = np.mean(svm_cv_scores)
svm_cv_scores
```

Find mean and standard deviation of difference:

```python
diffs = logreg_cv_scores - svm_cv_scores
diffs_mean = np.mean(diffs)
diffs_std = np.std(diffs)
diffs
```

Degrees of freedom: 28, Confidence: 90% --> Critical Value: 1.701 (from t-table)

```python
sig_diff = diffs_std/(30**0.5)*1.701
print("Interval: [{}, {}]".format(-1*sig_diff, sig_diff))
```

```python
print("Difference of means: ", diffs_mean)
```

The difference of the means is outisde of the interval, therefore we can say we 90% confidence that the logistic regression classifier is significantly better.

# Part 2: Business Analysis
## Lift chart:

```python
logreg.fit(data_rows, resp)
prob = logreg.predict_proba(data_rows)
pre = logreg.predict(data_rows)
```

```python
cols = {'Actual': resp, 'Predicted': pre, 'Probabilities': prob[:, 0]}
results = pd.DataFrame(cols)
  
results
```

```python
rand_order = results
rand_order['Cumulative_Pos'] = results['Actual'].cumsum()
rand_order
```

```python
sorted_order = results.sort_values('Probabilities')
sorted_order['Cumulative_Pos'] = sorted_order['Actual'].cumsum()
sorted_order
```

```python
import matplotlib.pyplot as plt
c_sent = np.arange(1, len(results) + 1)
plt.plot(c_sent, rand_order['Cumulative_Pos']/len(c_sent), label = "random")
plt.plot(c_sent, sorted_order['Cumulative_Pos']/len(c_sent), label = "sorted")
plt.xlabel('campaigns sent')
plt.ylabel('success rate')
plt.title('Lift')
plt.legend()
plt.show()
```

## Cost Benefit Analysis:

Cost to mail campaign: 50 cents, benefit from campaign: $20

```python
cb = pd.DataFrame({'Num_Campaigns': c_sent,'Benefit': sorted_order['Cumulative_Pos']*20})
cb['Cost'] = c_sent * .5
cb['Expected_Rev'] = cb['Benefit'] - cb['Cost']
cb.iloc[cb['Expected_Rev'].argmax()]
```
